```python
def corrplot(df, size = 18):
    corr = df.corr()
    fig, ax = plt.subplots(figsize = (size, size))
    w = ax.matshow(corr)
    cbar=fig.colorbar(w, orientation = 'vertical', shrink = .8, aspect = 15)
    cbar.ax.tick_params(labelsize = 20)
    ax.get_xaxis().tick_top()
    plt.xticks(range(len(corr.columns)), corr.columns,fontsize = 20, rotation = 75);
    plt.yticks(range(len(corr.columns)), corr.columns,fontsize = 20, rotation = 15);
```

```python
corrplot(wine)
```

```python
def scat(x,y,z):
    with plt.style.context('fivethirtyeight'):
        fig, ax = plt.subplots(figsize = (20, 15))
        ax.spines["top"].set_visible(False)
        ax.spines["right"].set_visible(False)
        ax.set_title('Wine Quality by alcohol and volatile acidity', fontsize = 40)
        ax.set_xlabel('volatile acidity', fontsize = 25)
        ax.set_ylabel('alcohol content', fontsize = 25)
        plt.xticks(fontsize = 20)
        plt.yticks( fontsize = 20)
        
        #ax.legend( wine[quality])
        scatplt = ax.scatter(x,y, s = 150, c = z, alpha = 1)
        cbaxes = fig.add_axes([0.4, .85, 0.2, .02]) 
        cb = fig.colorbar(scatplt,cax = cbaxes, orientation = 'horizontal', label = 'Wine Quality')
        
#cb = plt.colorbar(ax1, cax = cbaxes)  
    
    #cbar = fig.colorbar(wine['quality'], orientation = 'horizontal', vmin = 3, vmax = 8)
```
